```python
a = np.random.randint(1,10,(3,3))
b = np.random.randint(1,10,(3,3))
print(a)
print(b)
print(a+b)
print(a-b)
print(a/b)
```

```python
print(a.max())
print(a.mean())
print(a.min())
print(a.std()) #standard deviation 
print(a.sum())
```

```python
print(a.max(axis = 0)) #coloumn wise maximum value
print(a.max(axis = 1))  #row wise max value
```
