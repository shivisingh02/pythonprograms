```python
#random line plot
plt.figure(figsize = (15,6), dpi = 300) #dpi = dots per inch if you want high quality orint increase dpi
x= np.linspace(0 , 10 , 100)
y = np.random.randn(100)
_ =plt.plot(x ,y , linewidth = 3 ,color = 'yellow' , linestyle = ':' , marker = '*' , markersize = 10) 
#_underscore is used to collect garbage value
#marker is used to find spaces where x = y 
```

```python
#random bar plot
plt.figure(figsize = (15,4) , dpi = 300)
x = np.random.randint(0,10,25)
y = np.random.randint(0,10,25)
plt.grid(False)
_ = plt.bar(x ,y ,color = 'blue')
```
