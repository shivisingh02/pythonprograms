```python
pd.set_option('display.max_columns' , None)
```

```python
df = pd.read_excel('Canada.xlsx' , skiprows = 20, 
skipfooter = 2 , sheet_name = 'Canada by Citizenship')
df.head()
```

```python
#data cleaning
cols_to_drops = ['Type' ,'Coverage' , 'AREA' , 'REG' ,'DEV']
df.drop(columns = cols_to_drops , inplace = True)
```

```python
df
```

```python
print(df.columns.to_list())
```

```python
df.rename(
    columns = {
        'OdName' :'Country' ,
        'AreaName': 'Continent' ,
        'RegName' : 'Region' ,
        'DevName' : 'Status'
    },
    inplace = 'True')
```

```python
df.head(3)
```

```python
years = list(range(1980 , 2014))
print(years)
```

```python
df['Total'] = df[years].sum(axis = 1)
```

```python
df.head(3)
```

```python
#visualization
```

```python
df.set_index('Country' , inplace = True)
```

```python
df.sort_values(by = 1980 , ascending = False).head(25)[1980].plot(kind = 'bar' , figsize = (20 ,5) )
```

```python
df[1980].plot()
```

```python
df.sort_values(by = 2000 , ascending = False).head(5)[2000].plot(kind = 'bar' , figsize = (20 ,5) )
```

```python
df.sort_values(by = 'Total' , ascending = False).head(5)['Total'].plot(kind = 'bar' , figsize = (20 ,5) )
```
